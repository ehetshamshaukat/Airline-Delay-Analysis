# Data Analysis

## 1.Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
from plotly import offline
from plotly import graph_objs as go
```

### 2.Reading data

```python
dataset=pd.read_csv("/kaggle/input/airlines-dataset-to-predict-a-delay/Airlines.csv")
dataset.head(10)
```

### 3.Preprocessing

```python
dataset.isna().sum()
```

```python
dataset.dtypes
```

```python
dataset["Delay"].replace([0,1],["Delayed","Not Delayed"],inplace=True)
dataset["DayOfWeek"].replace([1,2,3,4,5,6,7],["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],inplace=True)
```

### Basic Analysis

* ### By Airline

```python
different_airline=dataset["Airline"].unique()
num_of_different_airline=dataset["Airline"].nunique()
print("Different airline name: ",str(different_airline))
print("Total number of airlines are: ",num_of_different_airline)
```

```python
airline=dataset["Airline"].value_counts()
airline=pd.DataFrame(airline).reset_index()
airline
```

```python
labels=airline["index"].index
values=airline["Airline"].values
trace0=go.Pie(values=values,labels=labels,hole=0.4,textinfo="label+percent")
layout=dict(title="Number of airline.",title_x=0.5)
data=[trace0]
fig=dict(layout=layout,data=data)
offline.iplot(fig)
```

```python
trace0=go.Bar(x=airline["index"],y=airline["Airline"])
data=[trace0]
layout=dict(title="Number of Airline",title_x=0.5,xaxis=dict(title="Name of Airlines"),yaxis=dict(title="Frequency"))
fig=dict(data=data,layout=layout)
offline.iplot(fig)
```

* ## By Time

```python
dataset.groupby(["Airline","Flight"]).agg(max_time=("Time","max"),min_time=("Time","min"),total_time=("Time","sum"),average_Time=("Time","mean")).reset_index()
```

```python
dataset.sort_values(by="Flight",ascending=False).groupby(["Airline","Flight"]).agg(max_Length=("Length","max"),min_Length=("Length","min"),total_Length=("Length","sum"),average_Length=("Length","mean")).reset_index()
```

* ## By Week

```python
num_of_day_week=dataset["DayOfWeek"].unique()
num_of_day_week
count_of_day_week=dataset["DayOfWeek"].value_counts(sort=True)
count_of_day_week
```

```python
day_and_airline=dataset.groupby(["DayOfWeek","Airline"]).agg(count_in_a_day=("DayOfWeek","count")).reset_index()
day_and_airline
```

```python
trace0=go.Bar(x=day_and_airline["DayOfWeek"],y=day_and_airline["count_in_a_day"],text=day_and_airline["Airline"],textposition="inside")
data=[trace0]
layout=dict(title="Days and Airline",title_x=0.5,xaxis=dict(title="Days"),yaxis=dict(title="Frequency"))
fig=dict(data=data,layout=layout)
offline.iplot(fig)
```

```python
days=dataset.groupby("DayOfWeek").agg(count_in_a_day=("DayOfWeek","count")).reset_index()
days
```

```python
values=days["count_in_a_day"]
labels=days["DayOfWeek"]
trace0=go.Pie(labels=labels,values=values,hole=0.4,textinfo="label+percent",textposition="inside",text=days["DayOfWeek"])
layout=dict(title="Number of flights overall(whole)",title_x=0.4)
data=[trace0]
fig=dict(data=data,layout=layout)
offline.iplot(fig)
```
